###Text file conversion

```python
import seaborn as sns
import pandas as pd
import numpy as np
df = pd.DataFrame({'File Size': ['200MB','400MB', '600MB', '800MB', '1000MB'],
                   'C' : [12, 20, 34, 45, 55],
                   'C++': [15, 25, 36, 50, 60],
                   'Java': [18, 30, 40, 55, 70],
                   'R': [20, 35, 45, 60, 80],
                   'Python': [25, 40, 53, 75, 100]})
df
```

```python
x  = [1, 2, 3, 4, 5]
y1 = [12, 20, 34, 45, 55]
y2 = [15, 25, 36, 50, 60]
y3 = [18, 30, 40, 55, 70]
y4 = [20, 35, 45, 60, 80]
y5 = [25, 40, 53, 75, 100]
graph = sns.lineplot(x=x, y=y1, label='C', linewidth=2)
sns.lineplot(x = x, y = y2, label='C++', linewidth=2)
sns.lineplot(x = x, y = y3, label='Java', linewidth=2)
sns.lineplot(x = x, y = y4, label='R', linewidth=2)
sns.lineplot(x = x, y = y5, label='Python', linewidth=2)
graph.set_xlabel('File Size(GB)')
graph.set_ylabel('Time Taken(in sec)')
graph.set_title('Performance Comparision of Different Programming Languages')
```
