```python
# (i) Replace all the values with NA in the dataset defined between [10, 60]. Print the count of number rows having missing values.

import random

num_rows = 100
num_columns = 30

dataset = []
missing_rows_count = 0

for _ in range(num_rows):
    row = [random.randint(1, 200) for _ in range(num_columns)]
    for i in range(num_columns):
        if 10 <= row[i] <= 60:
            row[i] = 'NA'
    dataset.append(row)
    
    if 'NA' in row:
        missing_rows_count += 1

print(f"Number of rows with missing values: {missing_rows_count}")

```

```python
# Normalize all the values in the dataset between 0 and 10.

import random
import numpy as np

num_rows = 100
num_columns = 30

dataset = []

for _ in range(num_rows):
    row = [random.randint(1, 200) for _ in range(num_columns)]
    dataset.append(row)

# Normalize dataset between 0 and 10
normalized_dataset = []
for row in dataset:
    min_value = min(row)
    max_value = max(row)
    normalized_row = [(value - min_value) / (max_value - min_value) * 10 for value in row]
    normalized_dataset.append(normalized_row)

# Printing the normalized dataset (optional)
for row in normalized_dataset:
    print(row)
```

```python
# (v) Replace all the values in the dataset with 1 if value <=0.5 else with 0.

import random

num_rows = 100
num_columns = 30

dataset = []

for _ in range(num_rows):
    row = [random.uniform(0, 1) for _ in range(num_columns)]
    dataset.append(row)

# Replace values in dataset
for i in range(num_rows):
    for j in range(num_columns):
        if dataset[i][j] <= 0.5:
            dataset[i][j] = 1
        else:
            dataset[i][j] = 0

# Printing the modified dataset (optional)
for row in dataset:
    print(row)
```

```python
# Apply following clustering algorithms; determine the optimal number of clusters and plot distance metric
#graph using each algorithm. (i) K-Mean clustering

import random
from sklearn.cluster import KMeans

num_rows = 500
num_columns = 10
num_clusters = 3  # Number of clusters to create

# Generate the dataset
dataset = []
for _ in range(num_rows):
    row = []
    for _ in range(4):
        value = random.uniform(-10, 10)
        row.append(value)
    for _ in range(4):
        value = random.uniform(10, 20)
        row.append(value)
    for _ in range(2):
        value = random.uniform(-100, 100)
        row.append(value)
    dataset.append(row)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(dataset)
labels = kmeans.labels_

# Printing the cluster labels (optional)
for label in labels:
    print(label)

```

```python
# ii) Hierarchical clustering

import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

num_rows = 500
num_columns = 10

dataset = []

for _ in range(num_rows):
    row = []
    # Columns 1 to 4: [-10, 10]
    for _ in range(4):
        value = random.uniform(-10, 10)
        row.append(value)
    # Columns 5 to 8: [10, 20]
    for _ in range(4):
        value = random.uniform(10, 20)
        row.append(value)
    # Columns 9 to 10: [-100, 100]
    for _ in range(2):
        value = random.uniform(-100, 100)
        row.append(value)
    dataset.append(row)

# Convert the dataset to a numpy array
data = np.array(dataset)

# Perform hierarchical clustering
linked = linkage(data, 'ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
# (ii) Plot histogram of each column in single graph.

import random
import numpy as np
import matplotlib.pyplot as plt

num_rows = 600
num_columns = 15

dataset = []

for _ in range(num_rows):
    row = [random.uniform(-100, 100) for _ in range(num_columns)]
    dataset.append(row)

# Convert the dataset to a numpy array
data = np.array(dataset)

# Create subplots for each column
fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(12, 12))

# Plot histograms for each column
for i, ax in enumerate(axes.flatten()):
    ax.hist(data[:, i], bins=30, alpha=0.7, color='blue', edgecolor='black')
    ax.set_title(f'Column {i+1}')
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

# Adjust spacing between subplots
fig.tight_layout()

# Show the plot
plt.show()
```

```python
# (i) Perform t-Test on each column.

import random
import numpy as np
from scipy.stats import ttest_1samp

num_rows = 500
num_columns = 5

dataset = []

for _ in range(num_rows):
    row = [random.uniform(5, 10) for _ in range(num_columns)]
    dataset.append(row)

# Convert the dataset to a numpy array
data = np.array(dataset)

# Perform t-test on each column
for column in range(num_columns):
    column_data = data[:, column]
    t_statistic, p_value = ttest_1samp(column_data, 7.5)  # Perform t-test against the null hypothesis mean of 7.5
    print(f"Column {column+1}: t-statistic = {t_statistic}, p-value = {p_value}")
```

```python
# (ii) Perform Wilcoxon Signed Rank Test on each column.

import random
import numpy as np
from scipy.stats import wilcoxon

num_rows = 500
num_columns = 5

dataset = []

for _ in range(num_rows):
    row = [random.uniform(5, 10) for _ in range(num_columns)]
    dataset.append(row)

# Convert the dataset to a numpy array
data = np.array(dataset)

# Perform Wilcoxon signed-rank test for each column
for i in range(num_columns):
    column_data = data[:, i]
    statistic, p_value = wilcoxon(column_data)
    print(f"Column {i+1}:")
    print(f"Statistic: {statistic}")
    print(f"P-value: {p_value}")
    print()
```

```python
# (iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4

import random
from scipy.stats import ttest_ind, ranksums

num_rows = 500
num_columns = 5

dataset = []

for _ in range(num_rows):
    row = [random.uniform(5, 10) for _ in range(num_columns)]
    dataset.append(row)

# Extract Column 3 and Column 4 from the dataset
column_3 = [row[2] for row in dataset]
column_4 = [row[3] for row in dataset]

# Perform two-sample t-test
t_statistic, p_value_ttest = ttest_ind(column_3, column_4)
print("Two-Sample t-test:")
print("t-statistic:", t_statistic)
print("p-value:", p_value_ttest)

# Perform Wilcoxon rank-sum test
z_statistic, p_value_wilcoxon = ranksums(column_3, column_4)
print("\nWilcoxon Rank-Sum Test:")
print("z-statistic:", z_statistic)
print("p-value:", p_value_wilcoxon)
```
